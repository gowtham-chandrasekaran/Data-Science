```python
def min_max_values(col):
    top = data[col].idxmax() #return id of max value under that column
    top_obs = pd.DataFrame(data.loc[top]) #Getting the row from that index to display
    
    bottom = data[col].idxmin() #return id of min value under that column
    bottom_obs = pd.DataFrame(data.loc[bottom]) #Getting the row from that index to display
    
    min_max_obs = pd.concat([top_obs,bottom_obs],axis = 1)
    return min_max_obs
```

```python
min_max_values('MPG (City)') #passing column name to above function to retrieve max and min value of that particular column and to get the total rows
```

From above, we can see that Honda Hybrid has the max mileage in city and Ford XLT has the minimum mileage in city

# Correlation using heatmap

```python
corr = cars_numerical.drop('Horsepower',axis=1).corr()
sns.heatmap(corr[(corr>=0.5)|(corr<=-0.5)], cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size":20}, square=True);
```

We are setting correlation threshold from 0.5 to -0.5
