# Data Cleaning

Let's check relation between Pclass and age

```python
plt.figure(figsize=(12,7))
sns.boxplot(x='Pclass',y='Age',data=data,palette='winter') #palette is for styling
```

We can see the median and mean age for each class. This can be used to fill the missing age values.

We are going to create a function to fill NA values of age with above mean age for each Pclass

```python
def missingage(cols):
    age = cols[0]
    pclass = cols[1]
    
    if pd.isnull(age):
        if pclass == 1:
            age = 37     #mean of class 1 from above figure
        if pclass == 2:
            age = 29    #mean of class 2 from above figure
        else:
            age = 24   #mean of class 3 from above figure
        return age
    else:
        return age
        
```

```python
data['Age'] = data[['Age','Pclass']].apply(missingage,axis=1)
```

```python
data.isnull().sum()
```

Age has no null values now.

```python
sns.heatmap(data.isnull(),cmap='viridis',yticklabels=False) #cmap is also for styling only
```

# Converting categorical into numerical columns

Now we have to change sex and embarked columns to numerical values using pd.get_dummies().
pandas.get_dummies() is used for data manipulation. It converts categorical data into dummy or indicator variables.

```python
sex = pd.get_dummies(data['Sex'],drop_first=True)
embarked = pd.get_dummies(data['Embarked'],drop_first=True)
```

It shows if a record is male or not, which is enough.

This has two columns as initially there were 3. Everytime one column will decrease if we give drop_first=True.

```python
data.drop(['Sex','Embarked','Name','Ticket'],inplace=True,axis=1)
```
