# Logistic Regression is used for classification problems.

# Let's look at an example of customers who bought health insurance

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Data analytics projects\ML Basics\Logistic Regression\insurance_data.csv")
```

```python
df
```

```python
plt.scatter(df.age,df.bought_insurance,marker="+",color="red")
```

0 means No and 1 means Yes.

We can infer that if the person is younger he is less likely to buy the insurance and vice versa.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['age']],df.bought_insurance, test_size=0.1)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

The model has predicted that for test values below 25,19 age will not buy and 58 will buy insurance which is what we also predicted.

```python
X_test
```

```python
model.score(X_test,y_test)
```

Our model is perfect 100%. Because of small size of dataset score is very high.

```python
model.predict_proba(X_test)
```

This function shows the probability of three ages 25,58,19 to buy the insurance. Left is Not buy, Right is Buy.
