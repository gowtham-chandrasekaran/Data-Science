Above we see a value like '1133 - 1384' which is a range. We can convert this as mean value and substitute it there.

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

In above two cells, we see string values and ranges.
In this project, we are only going to remove strings and have the ranges. Ranges can be converted to their mean

```python
df3 = df2.copy() #We are creating a copy for further cleaning
```

```python
def sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

In the above function, we are taking each value from the total_sqft column and trying to split with '-' to get a list of numbers only which is still a string. Next we convert the string to float and find average of the two and return it.
If the above operation does not perform, it will come down to just convert to float which is the case for all numbers that are entering into the function. This will throw an error if the x contains a string value other than '-' like 34.46Sq. Meter. For them, return none.

```python

df3['total_sqft'] = df3['total_sqft'].apply(sqft_to_num)
```
